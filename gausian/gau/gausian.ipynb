```python
import numpy as np
from scipy import ndimage
from skimage import io

# Load the image
image = io.imread("E:/Desktop/test-image/gau/9585931R.png")

# Define the Gaussian filter parameters
sigma = 4
truncate = 5

# Apply the filter
filtered_image = ndimage.gaussian_filter(image, sigma=sigma, truncate=truncate)

# Save the filtered image
io.imsave("E:/Desktop/test-image/gau/9585931Rd.png", filtered_image)
```

```python
from PIL import Image, ImageFilter

image = Image.open(r"E:/Desktop/test-image/gau/9919646L.png")

# Cropping the image
smol_image = image.crop((0, 0, 150, 150))

# Blurring on the cropped image
blurred_image = smol_image.filter(ImageFilter.GaussianBlur)

# Pasting the blurred image on the original image
image.paste(blurred_image, (0,0))

# Displaying the image
image.save('E:/Desktop/test-image/gau/9919646Lds.png')
```
